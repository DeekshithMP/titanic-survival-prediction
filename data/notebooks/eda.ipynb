```python
sns.countplot(x='Survived', data=df)
plt.title("Overall Survival Distribution")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
sns.barplot(x='Sex_male', y='Survived', data=df)
plt.title("Survival Rate by Gender")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Survival Rate")
plt.show()
```

```python
df['Pclass'] = 1
df.loc[df['Pclass_2'] == 1, 'Pclass'] = 2
df.loc[df['Pclass_3'] == 1, 'Pclass'] = 3

sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
embarked_cols = ['Embarked_Q', 'Embarked_S']
df['Embarked'] = 'C'
df.loc[df['Embarked_Q'] == 1, 'Embarked'] = 'Q'
df.loc[df['Embarked_S'] == 1, 'Embarked'] = 'S'

sns.barplot(x='Embarked', y='Survived', data=df)
plt.title("Survival Rate by Embarkation Port")
plt.xlabel("Embarked Port")
plt.ylabel("Survival Rate")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survival")
plt.xlabel("Survived")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Age',
    y='Fare',
    hue='Survived',
    style='Pclass',
    data=df
)
plt.title("Age vs Fare with Survival and Class")
plt.show()
```

```python
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title("Survival Rate by Family Size")
plt.xlabel("Family Size")
plt.ylabel("Survival Rate")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), cmap='coolwarm', annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()
```
